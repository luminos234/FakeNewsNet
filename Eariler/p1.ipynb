```python
query = """
SELECT 
    COUNT(*) AS count, 
    df.state_info, 
    states_df.alignment
FROM 
    df
JOIN 
    states_df ON df.state_info = states_df.state_info
WHERE 
    false_counts = 1.0 
GROUP BY 
    df.state_info, 
    states_df.alignment
ORDER BY 
    count DESC;
"""

result_df = pysqldf(query)
result_df
```
